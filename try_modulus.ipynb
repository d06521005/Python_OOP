# 挑選不同資料變成新table

```python
# 用 _.show_column()  秀出資料名稱跟對應編號
qd.show_column(['行政用電.csv', '設備用電.csv'])
```

```python
# 用qd.SELECT() 指令將不同資料變成新表格

# 用字典格式去讀{'table1': [column indexs], 'table2': [column indexs]}
select = {'設備用電.csv':[0,1,2,3,4,5,6],
          '行政用電.csv':[0]}
# 給一個新名稱(本例是list)
qd.SELECT(tablename='test', column_table = select) 
```

```python
# 用一樣的指令查看新 table
qd.pick_table['test']
```
